```python
import ultralytics
import wandb
from ultralytics import YOLO, settings
from IPython.display import Image, clear_output
!ultralytics checks
```

```python
wandb.login(relogin=True, force=True)
wandb.init(project="aerial_detection", name="40_epochs")
```

```python
model = YOLO('/kaggle/input/best-40/best-40.pt') 
model.train(epochs=40, batch=30,
            data='/kaggle/input/aerial-object-detection/data/data.yaml', 
            imgsz=1088, device=0, plots=True, optimizer='Adam', lr0=0.001, momentum=0.92859, save=True,
            weight_decay=0.00059, warmup_momentum=0.8, cls=0.40993, dfl=1.61501, box=7.5097, resume=True,
            fliplr=1)
```

```python
wandb.finish()
```

```python
results = model.val(data='/kaggle/input/aerial-object-detection/data.yaml',
                    save_json=True, device=0, split='val')
```

```python
results = model.val(data='/kaggle/input/aerial-object-detection/data.yaml',
                    save_json=True, device=0, split='test', imgsz=4000, batch=2)
```
